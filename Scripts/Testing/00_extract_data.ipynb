```python
# from os import listdir
# from os.path import isfile, join

import os
import shutil
```

```python
# dataverse_files is a folder where all folders are after unzipping
# unzipped is a folder to put everything after extracting from folders 
# both are removed in the end
folder = r"../../Data/optical-rs/dataverse_files"
destination_folder = r"../../Data/optical-rs/unzipped/"
subfolders = [f.path for f in os.scandir(folder) if f.is_dir()]
```

```python

for sub in subfolders:
    date = sub[-10:]
    for f in os.listdir(sub):
        src = os.path.join(sub, f)
        f = date + '_' + f
        #print(src) # full path
        #shutil.copy(src, destination_folder)
        dst = destination_folder + f
        print(dst)
        shutil.move(src, dst)
```

```python
for sub in subfolders:
    for f in os.listdir(sub):
        src = os.path.join(sub, f)
        shutil.copy(src, destination_folder)
```
