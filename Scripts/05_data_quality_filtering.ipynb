# Data Quality Filtering

**Purpose of script:**

Remove days of optical data that are of (too) bad data quality.
3) Do manual investigation of remaining days.

- In: opt data (all files)
- Out: file with list of days to be removed 

```python
import os
import pandas as pd
import xarray as xr
import rasterio
import matplotlib.pyplot as plt
```

```python
opt_path = r"../Data/optical-rs/"
```

```python
days_to_remove = set()
```

1) Remove all data from after 10th September of each year.

```python
#  iterate over all files in the optical-rs folder
for file in os.listdir(opt_path):
    # parse file names to string
    filename = str(file)
    date = file.split("_grain_diameter.tif")[0]
    # get date as pd.datetime
    date = pd.to_datetime(date, format="%Y-%m-%d")
    # append all dates that are after 10th September of each year to set
    if date.month == 9 and date.day > 10:
        days_to_remove.add(filename)

print(len(days_to_remove))
days_to_remove
```
