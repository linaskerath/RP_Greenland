```python
import xarray
import pandas as pd

from os import listdir
from os.path import isfile, join
```

```python
mw_path = r"../Data/microwave-rs/mw_interpolated/"
opt_path = r"../Data/optical-rs/unzipped/"

out_path = r"../Data/combined/dataframe_plain/"
```

```python
opt_files = [f for f in listdir(opt_path) if isfile(join(opt_path, f))]
```

```python
na_dict = {}
len = 3995569
for opt_file in opt_files:
    print(opt_file[:10])
    data_opt = xarray.open_dataarray(opt_path + opt_file).to_dataframe()
    na_dict[opt_file[:10]] = data_opt['band_data'].isna().sum() #/ len
```

```python
df = pd.DataFrame.from_dict(na_dict, orient='index').reset_index()
df.columns = ['date', 'na']
df['percentage'] = df['na'] / len
df['month-day'] = df['date'].str[5:]
df_filtered = df[df['month-day']<= '09-10']
```
